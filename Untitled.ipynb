# RSI Function
$$RSI = 100 - \frac{100}{1 + RS}$$
The relative strength indicator is an indicator for getting the values, which help to know, whether the prices oversold or overbought, so that they could be either bought or sold due to the two conditions above. 

```python
ta.RSI(df1.open.values, 14)
```

It contains the relative strength indicator for the values of df1.open but the first 14, because it begins from the number, which we define as a second parameter of our function talib.RSI()

## A new Array for studying rate of change
#### By choosing the first element as a reference and using the following formula 
$$ f_{i-1} = \frac{p_{i}}{p_{i-1}} - 1 $$
where i ={1,2,3,...}, $f_{i}$ describes the "Rate_Of_Change" column below and $p_{i}$ represents the elements of "open" column   

```python
#import pandas as pd
#df1 = pd.read_csv("dataexport.csv")  It is already defined.
df1['Rate_Of_Change'] = (df1['open'] / 168.37) - 1 # Formula for getting the rate of change
print df1
```

Now we have the rate of change coloumn.
