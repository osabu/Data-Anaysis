## A new Array for studying rate of change
#### By choosing the first element as a reference and using the following formula 
$$ f_{i-1} = \frac{p_{i}}{p_{i-1}} - 1 $$
where i ={1,2,3,...}, $f_{i}$ describes the "Rate_Of_Change" column below and $p_{i}$ represents the elements of "open" column   

```python
import pandas as pd
df1 = pd.read_csv("dataexport.csv")
df1['Rate_Of_Change'] = (df1['open'] / 168.37) - 1 # Formula for getting the rate of change
print df1
```

## Explorative Datenanalyse using Matplotlib

```python
import pandas as pd
import matplotlib.pyplot as mt
import numpy as np 
df1 = pd.read_csv("dataexport.csv")
df1['Rate_Of_Change'] = (df1['open'] / 168.37) - 1 # Formula for getting the rate of change 
mt.figure(figsize = (30 , 20))
mt.xlabel("Times in minutes") # title of the chart 
mt.title("Open prices and the rate of their change in minutes")
mt.plot(df1.index , df1.open)
mt.plot(df1.index , df1.Rate_Of_Change)
mt.legend(['Open prices' , 'Rate Of Change'] , loc = 0)
```
